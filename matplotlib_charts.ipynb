```
#barchart - Use case: github contribution
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Date = ["5/16/2020", "5/17/2020", "5/18/2020", "5/19/2020","5/20/2020",
        "5/21/2020", "5/22/2020", "5/23/2020", "5/24/2020", "5/25/2020", 
        "5/26/2020", "5/27/2020", "5/28/2020", "5/29/2020"]
contribution = [28,7,0,2,1,1,0,4,3,2,0,0,1,0]
#rotating text in xaxis
plt.xticks(rotation=75)

ax.bar(Date,contribution)
plt.show()
```

```
#matplot lib usecase2: favourite books - pie chart

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
csp_booknames = ['Mindset', 'Grit', 'Rich dad Poor dad', 'The Lean startup']
csp_ratings = [10, 7, 8, 9]
Shankar_booknames = ['Think and Grow rich', 'The Magic of Thinking big', 'Ponniyin selvan', 'Infinitheism']
Shankar_ratings = [9, 8, 10, 9.5]
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Mindset')

fig1, (ax1, ax2) = plt.subplots(1,2)

#ax1 = fig.add_axes([0, 0, .5, .5], aspect=1)
#ax2 = fig.add_axes([.5, .0, .5, .5], aspect=1)

patches, texts, autotexts = ax1.pie(csp_ratings, explode=explode, labels=csp_booknames, autopct='%1.1f%%',
        shadow=True, startangle=90, rotatelabels=True,pctdistance=0.5,labeldistance=1, radius = 1.2)
#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

for text in texts:
    text.set_color('maroon')
for autotext in autotexts:
    autotext.set_color('maroon')

ax2.pie(Shankar_ratings, explode=explode, labels=Shankar_booknames, autopct='%1.1f%%',
        shadow=True, startangle=90, rotatelabels=True,pctdistance=0.5,labeldistance=1, radius = 1.2)
#ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title('CSP'+"'"+'s book list')
ax2.set_title('Shankar'+"'"+'s book list',fontdict={'color':'Maroon','horizontalalignment':'center'})
plt.show()
```
